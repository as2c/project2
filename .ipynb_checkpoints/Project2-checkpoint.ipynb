# Q1.

```julia
using CairoMakie
using Statistics
function make_fig()
    fig = Figure()
    Axis(fig[1, 1])
    return fig
end
```

```julia
function find_poly(func, npoints, pmax, min, max)
    x = LinRange(min, max, npoints);
    fstar = func.(x)
    A = [x[i]^p for i in 1:npoints, p in 0:pmax];
    c = (A' * A) \ (A' * fstar);
    f(c, x) = sum(c[p + 1] * x^p for p in 0:pmax)
    yy = [f(c, xx) for xx in x]
    return x, fstar, yy
end
```

```julia
f(x) = sin(x)
min, max = 0, π / 2
pmax = 5
npoints = 1000
```

```julia
function plot_f_poly(x, yy, fstar, pmax)
    fig = make_fig()
    lines!(x, fstar, color=:red, linewidth=6)
    lines!(x, yy, color=:blue, linewidth=3)
    fig
end
```

```julia
function find_plot_poly(f, npoints, min, max, pmax)
    x, fstar, yy = find_poly(f, npoints, pmax, min, max);
    plot_f_poly(x, yy, fstar, pmax)
end
```

```julia
find_plot_poly(f, npoints, min, max, 0)
```

```julia
find_plot_poly(f, npoints, min, max, 1)
```

```julia
find_plot_poly(f, npoints, min, max, 2)
```

```julia
find_plot_poly(f, npoints, min, max, 3)
```

```julia
find_plot_poly(f, npoints, min, max, 5)
```

We can see that, in the range between [0, π/2] a third order polynomial is enough for approximating the sine function.

# Q2

```julia
f(x) = sin(x)
function calc_L2error(x, fstar, yy)
    return mean((yy .- fstar).^2)
end
```

```julia
function calc_errors(pmax_lev, f, npoints, min, max)
    pmaxs = collect(0:pmax_lev);
    errors = []
    for pmax in pmaxs
        x, fstar, yy = find_poly(f, npoints, pmax, min, max)
        push!(errors, calc_L2error(x, fstar, yy))
    end
    return Float32.(pmaxs), Float32.(errors)
end
```

```julia
function plot_L2errors(pmax_lev, lower_bound, f, npoints, min, max)
    pmaxs, errors = calc_errors(pmax_lev, f, npoints, min, max)
    fig = make_fig()
    scatter!(pmaxs[lower_bound:end], errors[lower_bound:end])
    fig
end
```

* Plotting all pmaxs from 0 to 20

```julia
pmax_lev = 20
plot_L2errors(pmax_lev, 1, f, npoints, min, max)
```

The error significantly decreases when 

* Plotting all pmaxs from 5 to 20

```julia
plot_L2errors(pmax_lev, 5)
```

The error for polynomial of order 5 is very low, so it is a good approximator of sine function in interval [0, π/2]

* Plotting all pmaxs from 8 to 20

```julia
plot_L2errors(pmax_lev, 8)
```

As you can see, we can't get an error less than e-16 because of the floating point error of the 32 bit Float values with which we were working. So, it's impossible to achieve an error of E < e-20 unless we work with values like double, which can store more bits in their decimals (more digits)

# Q3.

```julia
function find_odd_poly(func, npoints, pmax, min, max)
    x = LinRange(min, max, npoints);
    fstar = func.(x)
    A = [x[i]^(2*p + 1) for i in 1:npoints, p in 0:pmax];
    c = (A' * A) \ (A' * fstar);
    f(c, x) = sum(c[p + 1] * x^(2*p + 1) for p in 0:pmax)
    yy = [f(c, xx) for xx in x]
    return x, fstar, yy
end
```

```julia
function calc_odd_errors(pmax_lev, f, npoints, min, max)
    pmaxs = collect(0:pmax_lev);
    errors = []
    for pmax in pmaxs
        x, fstar, yy = find_odd_poly(f, npoints, pmax, min, max)
        push!(errors, calc_L2error(x, fstar, yy))
    end
    return Float32.(pmaxs), Float32.(errors)
end

function plot_odd_L2errors(pmax_lev, lower_bound, f, npoints, pmax, min, max)
    pmaxs, errors = calc_odd_errors(pmax_lev, f, npoints, pmax, min, max)
    fig = make_fig()
    scatter!(2 .* pmaxs[lower_bound:end] .+ 1, errors[lower_bound:end])
    fig
end
```

```julia
pmax_lev = 3
plot_odd_L2errors(pmax_lev, 1, f, npoints, pmax, min, max)
```

```julia
pmax_lev = 10
pmaxs_odd, errors_odd = calc_odd_errors(pmax_lev, f, npoints, pmax, min, max)
pmaxs, errors = calc_errors(pmax_lev, f, npoints, min, max);
```

```julia
function compare_poly_plots(lower_bound)
    fig = make_fig()
    lines!(pmaxs_odd[lower_bound:end], errors_odd[lower_bound:end], color=:cyan)
    lines!(pmaxs[lower_bound:end], errors[lower_bound:end], color=:red)
    fig
end
```

```julia
compare_poly_plots(1)
```

As you can see from the above plot, the error (the y axis) is much lower for the odd polynomials (the red plot) and converges faster as compared to when we keep all the polynomials (the cyan plot)
