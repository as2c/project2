# Q1.

```julia
using CairoMakie
using Statistics
# generic function to make figures with
function make_fig()
    fig = Figure()
    Axis(fig[1, 1])
    return fig
end
```

```julia
# find coefficients approximating the sine function with
# polynomials, i.e. all the c's
function find_coeffs(func, npoints, pmax, min, max)
    x = LinRange(min, max, npoints);
    fstar = func.(x)
    A = [x[i]^p for i in 1:npoints, p in 0:pmax];
    c = (A' * A) \ (A' * fstar);
    return c, x, fstar
end
```

```julia
# calculate the actual values of the sine function (fstar)
# and the approximated values (yy) in an interval delimited
# by min and max values. Here min = 0 and max = π/2
function find_poly(func, npoints, pmax, min, max)
    c, x, fstar = find_coeffs(func, npoints, pmax, min, max)
    f(c, x) = sum(c[p + 1] * x^p for p in 0:pmax)
    yy = [f(c, xx) for xx in x]
    return x, fstar, yy
end
```

```julia
#= initializing the sine function and min, max values,
as well as the pmax which is the maximum order of the
estimated polynomial, and npoints which is the number of
points chosen from the interval [0, π/2]
=#
f(x) = sin(x)
min, max = 0, π / 2
pmax = 5
npoints = 1000
```

```julia
#= plot the actual values of sine function (fstar) 
and the approximated values (yy)
=#
function plot_f_poly(x, yy, fstar)
    fig = make_fig()
    lines!(x, fstar, color=:red, linewidth=6)
    lines!(x, yy, color=:blue, linewidth=3)
    fig
end
```

```julia
#= this function uses the last two functions, i.e. 
it finds the polynomial and plots it.
=#
function find_plot_poly(f, npoints, min, max, pmax)
    x, fstar, yy = find_poly(f, npoints, pmax, min, max);
    plot_f_poly(x, yy, fstar)
end
```

Plotting fstar and approximated value (yy) for different
values of pmax:

```julia
find_plot_poly(f, npoints, min, max, 0)
```

```julia
find_plot_poly(f, npoints, min, max, 1)
```

```julia
find_plot_poly(f, npoints, min, max, 2)
```

```julia
find_plot_poly(f, npoints, min, max, 3)
```

```julia
find_plot_poly(f, npoints, min, max, 5)
```

We can see that, in the range between [0, π/2] a third order polynomial is enough for approximating the sine function.

# Q2

```julia
f(x) = sin(x)
function calc_L2error(x, fstar, yy)
    return mean((yy .- fstar).^2)
end
```

```julia
# calculate errors for all points and put them in an array
function calc_errors(pmax_lev, f, npoints, min, max)
    pmaxs = collect(0:pmax_lev);
    errors = []
    for pmax in pmaxs
        x, fstar, yy = find_poly(f, npoints, pmax, min, max)
        push!(errors, calc_L2error(x, fstar, yy))
    end
    return Float32.(pmaxs), Float32.(errors)
end
```

```julia
# plotting the errors
function plot_L2errors(pmax_lev, lower_bound, f, npoints, min, max)
    pmaxs, errors = calc_errors(pmax_lev, f, npoints, min, max)
    fig = make_fig()
    scatter!(pmaxs[lower_bound:end], errors[lower_bound:end])
    fig
end
```

* Plotting all pmaxs from 0 to 20

```julia
pmax_lev = 20
plot_L2errors(pmax_lev, 1, f, npoints, min, max)
```

The error significantly decreases when 

* Plotting all pmaxs from 5 to 20

```julia
plot_L2errors(pmax_lev, 5, f, npoints, min, max)
```

The error for polynomial of order 5 is very low, so it is a good approximator of sine function in interval [0, π/2]

* Plotting all pmaxs from 8 to 20

```julia
plot_L2errors(pmax_lev, 8, f, npoints, min, max)
```

As you can see, we can't get an error less than e-16 because of the floating point error of the 32 bit Float values with which we were working. So, it's impossible to achieve an error of E < e-20 unless we work with values like double, which can store more bits in their decimals (more digits)

# Q3.

```julia
# just like find_poly but only keeping odd powers
function find_odd_poly(func, npoints, pmax, min, max)
    x = LinRange(min, max, npoints);
    fstar = func.(x)
    A = [x[i]^(2*p + 1) for i in 1:npoints, p in 0:pmax];
    c = (A' * A) \ (A' * fstar);
    f(c, x) = sum(c[p + 1] * x^(2*p + 1) for p in 0:pmax)
    yy = [f(c, xx) for xx in x]
    return x, fstar, yy
end
```

```julia
# just like calc_errors in Q2 but only for odd powers
function calc_odd_errors(pmax_lev, f, npoints, min, max)
    pmaxs = collect(0:pmax_lev);
    errors = []
    for pmax in pmaxs
        x, fstar, yy = find_odd_poly(f, npoints, pmax, min, max)
        push!(errors, calc_L2error(x, fstar, yy))
    end
    return Float32.(pmaxs), Float32.(errors)
end
# just like plot_L2errors function in Q2 but for odd powers
function plot_odd_L2errors(pmax_lev, lower_bound, f, npoints, pmax, min, max)
    pmaxs, errors = calc_odd_errors(pmax_lev, f, npoints, min, max)
    fig = make_fig()
    scatter!(2 .* pmaxs[lower_bound:end] .+ 1, errors[lower_bound:end])
    fig
end
```

```julia
pmax_lev = 3
plot_odd_L2errors(pmax_lev, 1, f, npoints, pmax, min, max)
```

```julia
pmax_lev = 10
pmaxs_odd, errors_odd = calc_odd_errors(pmax_lev, f, npoints, min, max)
pmaxs, errors = calc_errors(pmax_lev, f, npoints, min, max);
```

```julia
#= comparing the errors of approximating by using 
only the odd power polynomials with approximation using all powers.
=#
function compare_poly_plots(lower_bound)
    fig = make_fig()
    lines!(pmaxs_odd[lower_bound:end], errors_odd[lower_bound:end], color=:cyan)
    lines!(pmaxs[lower_bound:end], errors[lower_bound:end], color=:red)
    fig
end
```

```julia
compare_poly_plots(1)
```

As you can see from the above plot, the error (the y axis) is much lower for the odd polynomials (the red plot) and converges faster as compared to when we keep all the polynomials (the cyan plot)

## 4.a)

f(c, x) = ∑ₚ₌₀ᵖᵐᵃˣ cₚ xᵖ → f'(c, x) = ∑ₚ₌₁ᵖᵐᵃˣ⁻¹ p × cₚ xᵖ<br>
but at the same time we have:<br>
g(d, x) = ∑ₚ₌₁ᵖᵐᵃˣ dₚ xᵖ<br>
therefore, by matching orders:<br>
dₚ = p × cₚ<br>
This means that the matrix D is of the following form:<br>
| 0 c₁ 0 0  ...|<br>
| 0 0 2c₂ 0 ...|<br>
| 0 0 0 3c₃ ...|<br>
| ...     |<br>

And we have |d> = D |c>

## 4.b)

```julia
# using SparseArrays
using LinearAlgebra
```

```julia
# based on 4.a) this function calculates D matrix
function calc_D(pmax)
    D = Float32.(diagm([i for i in 1:pmax]))
    print("D matrix:\n")
    display(D)
    return D
end

# This function finds |d> = D |c>
function find_d(func, npoints, pmax, min, max)
    c, x, fstar = find_coeffs(func, npoints, pmax, min, max)
    deleteat!(c, 1)
    return x, calc_D(pmax) * c
end
```

```julia
# This function calculates the derivative from the 
# coefficients |d> that we found.
function calc_deri_D(func, pmax, npoints, min, max)
    x, d = find_d(func, npoints, pmax, min, max)
    print("derivatie coefficients, d:")
    display(d)
    A = [x[i]^(p - 1) for i in 1:npoints, p in 1:pmax];
    return A * d
end
```

```julia
# Plots the derivatives calculated with D matrix and
# by approximating cos(x) using Q2 functions.
function plot_Dderi_vs_fprime(f_prime, npoints, min, max, pmax)
    deri = calc_deri_D(f, pmax, npoints, min, max)
    x, f_prime_star, yy_prime = find_poly(f_prime, npoints, pmax - 1, min, max);
    plot_f_poly(x, deri, yy_prime)
end
f_prime(x) = cos(x)
# d = find_d(f, npoints, pmax1, min, max)[2]
```
